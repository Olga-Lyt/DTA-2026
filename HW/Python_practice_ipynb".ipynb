Перше число: 10

### Завдання 18. Трикутник ⭐⭐⭐

Запитайте у користувача три довжини сторін і визначте:
1. Чи можна з них скласти трикутник? (правило: сума будь-яких двох сторін має бути більшою за третю)
2. Якщо можна — який це трикутник:
   - рівносторонній (всі сторони рівні)
   - рівнобедрений (дві сторони рівні)
   - різносторонній (всі різні)

> 💡 **Підказка:** Спочатку перевірте, чи трикутник взагалі існує. Потім вже визначайте тип через порівняння сторін.

````python
a = float(input("Введіть довжину першої сторони: "))
b = float(input("Введіть довжину другої сторони: "))
c = float(input("Введіть довжину третьої сторони: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    print("З цих сторін можна скласти трикутник.")
    if a == b and b == c:
        print("Це рівносторонній трикутник.")
    elif a == b or b == c or a == c:
        print("Це рівнобедрений трикутник.")
    else:
        print("Це різносторонній трикутник.")
else:
    print("З цих сторін не можна скласти трикутник.")
````

### Завдання 24. Факторіал ⭐⭐

Запитайте у користувача число `n` і обчисліть його факторіал (`n!`).

**Нагадування:** факторіал — це добуток всіх чисел від 1 до `n`.  
Наприклад, `5! = 1 * 2 * 3 * 4 * 5 = 120`.

> 💡 **Підказка:** Створіть змінну `result = 1`, а потім у циклі множте її на кожне число.

````python
n = int(input("Введіть число для обчислення факторіалу: "))

if n < 0:
    print("Факторіал не визначений для від'ємних чисел.")
elif n == 0:
    print("Факторіал 0 дорівнює 1.")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print(f"Факторіал числа {n} дорівнює {factorial}.")
````

### Завдання 25. Степінь числа ⭐⭐

Запитайте у користувача два числа: основу `a` і степінь `n`.
Обчисліть `a` в степені `n`, НЕ використовуючи оператор `**`.

**Приклад:** `a = 3, n = 4` → `3 * 3 * 3 * 3 = 81`

> 💡 **Підказка:** Створіть змінну `result = 1` і помножте її на `a` рівно `n` разів.

````python
a = float(input("Введіть основу (число a): "))
n = int(input("Введіть степінь (ціле число n): "))

result = 1

if n < 0:
    print("Для цього завдання степінь має бути невід'ємним цілим числом.")
elif n == 0:
    result = 1
else:
    for _ in range(n):
        result *= a

print(f"Число {a} в степені {n} дорівнює {result}")
````

### Завдання 26. Сума цифр числа ⭐⭐⭐

Запитайте у користувача число і знайдіть суму його цифр.

**Приклад:** число 1234 → сума 1+2+3+4 = 10

> 💡 **Підказка:** Найпростіший спосіб: перетворити число на рядок через `str()`, а потім перебрати символи у циклі і додавати `int(символ)` до суми.

````python
num_str = input("Введіть число: ")

total_sum_digits = 0
for digit_char in num_str:
    if digit_char.isdigit():
        total_sum_digits += int(digit_char)
    else:
        print("Будь ласка, вводьте лише цифри.")
        total_sum_digits = -1
        break

if total_sum_digits != -1:
    print(f"Сума цифр числа {num_str} дорівнює {total_sum_digits}.")
````

Виведіть всі прості числа від 2 до 50.

**Нагадування:** просте число — це число, яке ділиться тільки на 1 і саме на себе.

> 💡 **Підказка:** Для кожного числа `n` від 2 до 50 перевіряйте, чи має воно дільники від 2 до n-1. Можна використати змінну-прапорець `is_prime = True`.

````python
# Ваш код для завдання 27:
for num in range(2, 51):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
````

````python
num = int(input("Введіть додатнє ціле число: "))

if num < 0:
    print("Будь ласка, введіть додатнє ціле число.")
elif num == 0:
    print("Сума цифр числа 0 дорівнює 0.")
else:
    total_sum = 0
    temp_num = num
    while temp_num > 0:
        digit = temp_num % 10
        total_sum += digit
        temp_num = temp_num // 10
    print(f"Сума цифр числа {num} дорівнює {total_sum}.")
````

````python
a = int(input("Введіть перше додатнє число: "))
b = int(input("Введіть друге додатнє число: "))

if a <= 0 or b <= 0:
    print("Будь ласка, введіть додатні числа.")
else:
    while b != 0:
        a, b = b, a % b
    print(f"Найбільший спільний дільник: {a}")
````

````python
numbers = [42, 17, 89, 3, 56, 91, 24]
n = len(numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print(f"Відсортований список: {numbers}")
````

### Завдання 58. Шифр Цезаря ⭐⭐⭐

Запитайте у користувача слово (українською або англійською) і число `shift`.
Зашифруйте слово: кожну літеру замініть на ту, що стоїть на `shift` позицій далі в алфавіті.

**Приклад:** слово `abc`, `shift = 2` → `cde`.

> 💡 **Підказка:** Використайте функції `ord(letter)` (повертає код символу) і `chr(num)` (повертає символ за кодом). Для кожної літери: новий код = ord(letter) + shift. Для простоти можна не обробляти переходи через кінець алфавіту.

````python
word = input("Введіть слово для шифрування: ")
shift = int(input("Введіть зсув (число): "))

encrypted_word = ""
for char in word:
    if 'a' <= char <= 'z':
        start = ord('a')
        encrypted_char_code = (ord(char) - start + shift) % 26 + start
        encrypted_word += chr(encrypted_char_code)
    elif 'A' <= char <= 'Z':
        start = ord('A')
        encrypted_char_code = (ord(char) - start + shift) % 26 + start
        encrypted_word += chr(encrypted_char_code)
    elif 'а' <= char <= 'я':
        encrypted_word += chr(ord(char) + shift)
    elif 'А' <= char <= 'Я':
        encrypted_word += chr(ord(char) + shift)
    else:
        encrypted_word += char

print(f"Зашифроване слово: {encrypted_word}")
````

- Підрахуйте кількість спроб і виведіть її в кінці

````python
secret = 42
attempts = 0
guessed = False

print("Я загадав число від 1 до 100. Спробуй вгадати!")

while not guessed:
    try:
        guess = int(input("Введіть ваше число: "))
        attempts += 1

        if guess < secret:
            print("Більше")
        elif guess > secret:
            print("Менше")
        else:
            print(f"Вітаю! Ви вгадали число {secret} за {attempts} спроб!")
            guessed = True
    except ValueError:
        print("Будь ласка, введіть ціле число.")
````

Дано рядок з реченням. Виконайте такий аналіз і виведіть результати:
1. Кількість літер (без пробілів)
2. Кількість слів
3. Кількість голосних літер
4. Найдовше слово

> 💡 **Підказка:** Метод `.split()` розбиває рядок на список слів. Наприклад, `'hello world'.split()` повертає `['hello', 'world']`.

````python
text = "програмування на python це цікаво і корисно"

letter_count = 0
for char in text:
    if char.isalpha():
        letter_count += 1

words = text.split()
word_count = len(words)

vowels = "аеиіоуяюєїaeiou"
vowel_in_text_count = 0
for char in text.lower():
    if char in vowels:
        vowel_in_text_count += 1

longest_word = ""
if words:
    longest_word = words[0]
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word

print(f"Текст для аналізу: '{text}'")
print(f"1. Кількість літер (без пробілів): {letter_count}")
print(f"2. Кількість слів: {word_count}")
print(f"3. Кількість голосних літер: {vowel_in_text_count}")
print(f"4. Найдовше слово: '{longest_word}'")
````

### Завдання 61. Магазин: статистика продажів ⭐⭐⭐

Дано список продажів за тиждень (по одному числу на день, починаючи з понеділка).
Виведіть:
1. Загальну виручку за тиждень
2. Середню виручку на день
3. Найкращий день (з найбільшою виручкою) — назвою (Понеділок, Вівторок...)
4. Кількість днів, коли виручка була вищою за середню

> 💡 **Підказка:** Створіть список з назвами днів: `days = ["Понеділок", "Вівторок", ...]`. Індекс найкращого дня знайдіть у циклі (або через `sales.index(max(sales))`).

````python
sales = [3200, 2800, 4100, 3500, 5200, 7800, 6500]
days_of_week = ["Понеділок", "Вівторок", "Середа", "Четвер", "П'ятниця", "Субота", "Неділя"]

total_revenue = sum(sales)
print(f"1. Загальна виручка за тиждень: {total_revenue} грн")

average_revenue = total_revenue / len(sales)
print(f"2. Середня виручка на день: {average_revenue:.2f} грн")

max_sales = 0
best_day_index = -1
for i in range(len(sales)):
    if sales[i] > max_sales:
        max_sales = sales[i]
        best_day_index = i

if best_day_index != -1:
    print(f"3. Найкращий день: {days_of_week[best_day_index]} ({max_sales} грн)")
else:
    print("3. Не вдалося визначити найкращий день.")

days_above_average = 0
for s in sales:
    if s > average_revenue:
        days_above_average += 1
print(f"4. Кількість днів з виручкою вищою за середню: {days_above_average}")
````

### Завдання 62. Турнірна таблиця ⭐⭐⭐

Дано два списки: команди та їхні очки (відповідні позиції).
Виведіть турнірну таблицю, ВІДСОРТОВАНУ за спаданням очок, у такому форматі:
```
1. Шахтар — 67 очок
2. Динамо — 64 очки
...
```

> 💡 **Підказка:** Найпростіший спосіб: у циклі знаходьте команду з максимальними очками, виводьте її, потім видаляйте цю пару (команду і очки) з обох списків. Повторюйте, поки списки не спорожніють.

````python
teams = ["Динамо", "Шахтар", "Зоря", "Дніпро", "Ворскла", "Олександрія"]
points = [64, 67, 52, 48, 39, 45]
print("Турнірна таблиця:")

current_teams = list(teams)
current_points = list(points)

sorted_table_data = []

while current_teams:
    max_points_current_iter = -1
    max_index_current_iter = -1

    for i in range(len(current_points)):
        if current_points[i] > max_points_current_iter:
            max_points_current_iter = current_points[i]
            max_index_current_iter = i

    if max_index_current_iter != -1:
        team_name = current_teams.pop(max_index_current_iter)
        team_points = current_points.pop(max_index_current_iter)
        sorted_table_data.append((team_name, team_points))

for rank, (team, score) in enumerate(sorted_table_data, 1):
    print(f"{rank}. {team} — {score} очок")
````
