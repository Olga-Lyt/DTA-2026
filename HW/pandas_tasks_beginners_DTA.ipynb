## Підготовка: Генерація тестових даних

Запустіть цю комірку перед початком роботи. Не змінюйте код.

````python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 200

categories = ['Електроніка', 'Одяг', 'Книги', 'Продукти', 'Спорт']
cities = ['Київ', 'Харків', 'Одеса', 'Дніпро', 'Львів']
payment_methods = ['Картка', 'Готівка', 'Онлайн']

df = pd.DataFrame({
    'order_id':        range(1001, 1001 + n),
    'date':            pd.date_range('2023-01-01', periods=n, freq='D')[:n],
    'category':        np.random.choice(categories, n),
    'product':         ['Товар_' + str(i) for i in np.random.randint(1, 51, n)],
    'city':            np.random.choice(cities, n),
    'quantity':        np.random.randint(1, 10, n),
    'price':           np.round(np.random.uniform(50, 5000, n), 2),
    'payment_method':  np.random.choice(payment_methods, n),
    'rating':          np.random.choice([1, 2, 3, 4, 5, np.nan], n, p=[0.05, 0.1, 0.15, 0.3, 0.3, 0.1]),
    'is_returned':     np.random.choice([True, False], n, p=[0.1, 0.9]),
})

# Додаємо кілька пропущених значень
df.loc[np.random.choice(df.index, 10, replace=False), 'city'] = np.nan
df.loc[np.random.choice(df.index, 5, replace=False), 'price'] = np.nan

# Розраховуємо загальну суму
df['total'] = df['quantity'] * df['price']

print('Дані успішно згенеровано. Рядків:', len(df))
print('Колонки:', list(df.columns))
````

## Завдання 14 (Підвищена складність). Накопичувальні показники

1. Відсортуйте датафрейм за датою.
2. Додайте колонку `cumulative_total` — накопичувальний обсяг продажів з початку року (`cumsum()`).
3. Визначте, в який день було досягнуто 50% від загального річного обсягу продажів.

````python
# Ваш код тут

df_sorted = df.sort_values('date').copy()

df_sorted['cumulative_total'] = df_sorted['total'].cumsum()

total_annual_sales = df_sorted['total'].sum()
half_annual_sales = total_annual_sales * 0.5

day_50_percent_reached = df_sorted[df_sorted['cumulative_total'] >= half_annual_sales]['date'].iloc[0]

display(df_sorted.head())
print(f"\nЗагальний річний обсяг продажів: {total_annual_sales:.2f}")
print(f"50% від річного обсягу продажів ({half_annual_sales:.2f}) було досягнуто: {day_50_percent_reached.strftime('%Y-%m-%d')}")
````

## Завдання 15 (Підвищена складність). apply() та lambda

1. Використайте `apply()` з lambda-функцією для створення колонки `total_with_tax`, де ПДВ становить 20%, але лише для замовлень на суму понад 1000 (решта — без змін).
2. Напишіть функцію `classify_rating(r)`, що повертає `'Поганий'` (1–2), `'Нейтральний'` (3), `'Хороший'` (4–5), `'Без оцінки'` (NaN). Застосуйте її до колонки `rating`.

````python
# Ваш код тут

df['total_with_tax'] = df['total'].apply(lambda x: x * 1.20 if x > 1000 else x)

def classify_rating(r):
    if pd.isna(r):
        return 'Без оцінки'
    elif r >= 4:
        return 'Хороший'
    elif r >= 3:
        return 'Нейтральний'
    else:
        return 'Поганий'

df['rating_category'] = df['rating'].apply(classify_rating)

display(df[['total', 'total_with_tax', 'rating', 'rating_category']].head())

good_orders = (df['rating_category'] == 'Хороший').sum()
total_orders = len(df)

share_good = good_orders / total_orders * 100

print(f"Кількість замовлень з хорошим рейтингом: {good_orders}")
print(f"Частка замовлень з хорошим рейтингом: {share_good:.2f}%")
````

**Висновок:** *(Яка частка замовлень має хороший рейтинг? Коли apply кращий за vectorized-операції?)*  
Кількість замовлень з хорошим рейтингом 58%. Аpply кращий за vectorized-операції у випадках, коли потрібні складні розрахунки. В інших випадках доцільніше використовувати vectorized-операції, тому що вони працюють швидше та будуть більш ефективними.

````python
# Ваш код тут

df_discounts = pd.DataFrame({
    'category':     ['Електроніка', 'Одяг', 'Книги', 'Продукти', 'Спорт'],
    'discount_pct': [0.10, 0.20, 0.05, 0.00, 0.15]
})

df['total'] = df['quantity'] * df['price']

df_merged = pd.merge(
    df,
    df_discounts,
    on='category',
    how='left'
)

df_merged['final_total'] = (
    df_merged['total'] * (1 - df_merged['discount_pct'])
).round(2)

df_merged['discount_amount'] = (
    df_merged['total'] - df_merged['final_total']
).round(2)

discount_by_category = (
    df_merged
    .groupby('category')['discount_amount']
    .sum()
    .sort_values(ascending=False)
)

top_category = discount_by_category.idxmax()
top_discount = discount_by_category.max()

print("Сумарна знижка по категоріях:")
print(discount_by_category)

print(
    f"\nНайбільшу сумарну знижку отримала категорія "
    f"'{top_category}' — {top_discount:.2f}"
)

display(
    df_merged[
        [
            'order_id',
            'category',
            'total',
            'discount_pct',
            'discount_amount',
            'final_total'
        ]
    ].head()
)
````

````python
# Ваш код тут
df_merged['city_avg_total'] = (
    df_merged
    .groupby('city')['total']
    .transform('mean')
    .round(2)
)

df_merged['above_city_avg'] = (
    df_merged['total'] > df_merged['city_avg_total']
)

city_share = (
    df_merged
    .groupby('city')['above_city_avg']
    .mean()
    .mul(100)
    .round(2)
    .sort_values(ascending=False)
)

print("Частка замовлень вище середнього по місту (%):")
print(city_share)

display(
    df_merged[
        [
            'order_id',
            'city',
            'total',
            'city_avg_total',
            'above_city_avg'
        ]
    ].head()
)
````

Тransform доцільно використовувати тоді, коли потрібно створити нову колонку і порахувати в ній значення. Найбільша частка великих замовлень у Харкові.

## Завдання 18 (Складне). Ковзне середнє та аномалії

1. Згрупуйте дані за місяцем та розрахуйте щоденну виручку.
2. Розрахуйте 7-денне ковзне середнє виручки (`rolling(7).mean()`).
3. Визначте дні-аномалії: дні, коли виручка відхилилася від ковзного середнього більш ніж на 2 стандартних відхилення.

````python
# Ваш код тут
daily_revenue = (
    df_merged
    .groupby('date')['total']
    .sum()
    .reset_index()
    .sort_values('date')
)

daily_revenue['month'] = daily_revenue['date'].dt.to_period('M')

daily_revenue['rolling_mean_7d'] = (
    daily_revenue['total']
    .rolling(window=7)
    .mean()
)

daily_revenue['rolling_std_7d'] = (
    daily_revenue['total']
    .rolling(window=7)
    .std()
)

daily_revenue['upper_bound'] = (
    daily_revenue['rolling_mean_7d']
    + 2 * daily_revenue['rolling_std_7d']
)

daily_revenue['lower_bound'] = (
    daily_revenue['rolling_mean_7d']
    - 2 * daily_revenue['rolling_std_7d']
)

daily_revenue['is_anomaly'] = (
    (daily_revenue['total'] > daily_revenue['upper_bound']) |
    (daily_revenue['total'] < daily_revenue['lower_bound'])
)

anomalies = daily_revenue[daily_revenue['is_anomaly']]

print(f"Кількість аномальних днів: {len(anomalies)}")

display(
    anomalies[
        [
            'date',
            'total',
            'rolling_mean_7d',
            'upper_bound',
            'lower_bound'
        ]
    ]
)
````

**Висновок:** *(Скільки аномальних днів виявлено? Що може бути причиною різких відхилень виручки?)*  
Виявлено три аномальних днів. Причиною різких відхилень виручки може бути безліч факторів, наприклад: знижки, свята, рекламні кампанії, оптові покупці та ін.
