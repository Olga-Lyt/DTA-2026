# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

**Приклад:** для слова `програма` → `{'п': 1, 'р': 2, 'о': 1, 'г': 1, 'а': 2, 'м': 1}`

> 💡 **Підказка:** Перебирайте літери у циклі. Для кожної літери: якщо вона вже є у словнику — збільшіть значення на 1, якщо немає — додайте з 1.

````python
word = input("Введіть слово: ").lower()
letter_counts = {}

for char in word:
    if char.isalpha(): # Consider only alphabetic characters
        letter_counts[char] = letter_counts.get(char, 0) + 1

print(letter_counts)
````

Дано словник з товарами магазину, де значення — це інший словник з ціною і кількістю на складі. Виконайте:
1. Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).
2. Виведіть назви товарів, яких залишилось менше 20 штук.
3. Виведіть найдорожчий товар.

````python
inventory = {
    "Ноутбук":   {"price": 18500, "stock": 12},
    "Телефон":   {"price": 9200,  "stock": 35},
    "Планшет":   {"price": 7800,  "stock": 8},
    "Навушники": {"price": 850,   "stock": 120},
    "Клавіатура":{"price": 1200,  "stock": 45},
    "Мишка":     {"price": 450,   "stock": 18}
}
# Ваш код:

total_inventory_value = 0
for item, details in inventory.items():
    total_inventory_value += details['price'] * details['stock']
print(f"Загальна вартість всього товару на складі: {total_inventory_value} грн")

low_stock_items = []
print("\nТовари, яких залишилось менше 20 штук:")
for item, details in inventory.items():
    if details['stock'] < 20:
        low_stock_items.append(item)
        print(f"{item} ({details['stock']} шт)")

most_expensive_item = ""
max_price = 0
for item, details in inventory.items():
    if details['price'] > max_price:
        max_price = details['price']
        most_expensive_item = item
print(f"\nНайдорожчий товар: {most_expensive_item} ({max_price} грн)")
````

Дано список покупок, де кожна покупка — це словник з полями `product`, `category`, `price`. Створіть новий словник, де ключ — категорія, а значення — загальна виручка по цій категорії.

> 💡 **Підказка:** Створіть порожній словник `revenue = {}`. У циклі для кожної покупки: якщо її категорія вже є у словнику — додайте ціну до існуючого значення, якщо немає — створіть запис з ціною як початковим значенням.

````python
purchases = [
    {"product": "Хліб",       "category": "Продукти",  "price": 25},
    {"product": "Молоко",     "category": "Продукти",  "price": 38},
    {"product": "Шампунь",    "category": "Гігієна",   "price": 120},
    {"product": "Зубна паста","category": "Гігієна",   "price": 65},
    {"product": "Сир",        "category": "Продукти",  "price": 180},
    {"product": "Книга",      "category": "Розваги",   "price": 220},
    {"product": "Мило",       "category": "Гігієна",   "price": 35},
    {"product": "Фільм",      "category": "Розваги",   "price": 150},
    {"product": "Ковбаса",    "category": "Продукти",  "price": 195}
]
# Ваш код:

revenue = {}
for purchase in purchases:
    category = purchase['category']
    price = purchase['price']
    revenue[category] = revenue.get(category, 0) + price

print("Загальна виручка по категоріях:")
for category, total_price in revenue.items():
    print(f"{category}: {total_price} грн")
````

### Завдання 22. Телефонна книга з пошуком ⭐⭐⭐

Створіть словник-телефонну книгу з 5 контактами. Напишіть програму, яка у циклі `while`:
1. Запитує у користувача дію: `пошук`, `додати`, `видалити`, `вихід`
2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
phonebook = {
    "Олена": "+380501111111",
    "Андрій": "+380672222222",
    "Марія": "+380633333333",
    "Богдан": "+380994444444",
    "Софія": "+380965555555"
}

print("Ваша телефонна книга:", phonebook)

while True:
    action = input("\nВиберіть дію (пошук, додати, видалити, вихід): ").lower().strip()

    if action == "вихід":
        print("Дякуємо за використання телефонної книги!")
        break

    elif action == "пошук":
        name = input("Введіть ім'я для пошуку: ").strip()
        if name in phonebook:
            print(f"Номер телефону {name}: {phonebook[name]}")
        else:
            print(f"Контакт '{name}' не знайдено.")

    elif action == "додати":
        name = input("Введіть ім'я нового контакту: ").strip()
        if name in phonebook:
            print(f"Контакт '{name}' вже існує. Оновіть номер телефону, якщо потрібно.")
        phone_number = input(f"Введіть номер телефону для {name}: ").strip()
        phonebook[name] = phone_number
        print(f"Контакт '{name}' додано/оновлено.")

    elif action == "видалити":
        name = input("Введіть ім'я контакту для видалення: ").strip()
        if name in phonebook:
            del phonebook[name]
            print(f"Контакт '{name}' видалено.")
        else:
            print(f"Контакт '{name}' не знайдено.")

    else:
        print("Невірна дія. Спробуйте ще раз.")

    print("Поточна телефонна книга:", phonebook)
````

### Завдання 26. Калькулятор з історією ⭐⭐

У циклі `while` запитуйте у користувача два числа і операцію (`+`, `-`, `*`, `/`). Обчислюйте результат і зберігайте кожен у списку `history`. Цикл завершується, коли користувач введе `вихід` замість операції. У кінці виведіть усю історію обчислень.

````python
# Ваш код для завдання 26:
history = []

while True:
    try:
        num1_str = input("Введіть перше число (або 'вихід' для завершення): ")
        if num1_str.lower() == 'вихід':
            break
        num1 = float(num1_str)

        num2_str = input("Введіть друге число (або 'вихід' для завершення): ")
        if num2_str.lower() == 'вихід':
            break
        num2 = float(num2_str)

        operation = input("Введіть операцію (+, -, *, /) (або 'вихід' для завершення): ")
        if operation.lower() == 'вихід':
            break

        result = None
        if operation == '+':
            result = num1 + num2
        elif operation == '-':
            result = num1 - num2
        elif operation == '*':
            result = num1 * num2
        elif operation == '/':
            if num2 != 0:
                result = num1 / num2
            else:
                print("Помилка: Ділення на нуль неможливе!")
                continue
        else:
            print("Невідома операція. Спробуйте ще раз.")
            continue

        print(f"Результат: {result}")
        history.append({"num1": num1, "num2": num2, "operation": operation, "result": result})

    except ValueError:
        print("Неправильний ввід. Будь ласка, введіть число.")
    except Exception as e:
        print(f"Виникла непередбачена помилка: {e}")

print("\nІсторія обчислень:")
if history:
    for entry in history:
        print(f"{entry['num1']} {entry['operation']} {entry['num2']} = {entry['result']}")
else:
    print("Історія порожня.")
````

### Завдання 35. Рейтинг команд ⭐⭐

Дано список матчів. Кожен матч — це словник `{"team_a": ..., "team_b": ..., "score_a": ..., "score_b": ...}`. Створіть таблицю очок: за перемогу — 3 очка, за нічию — 1, за поразку — 0. Виведіть команди у порядку від найбільшої до найменшої кількості очок.

> 💡 **Підказка:** Створіть словник `points = {}`. У циклі по матчах: визначайте переможця і додавайте очки. Не забудьте обробити випадок, коли команди ще немає у словнику.

````python
matches = [
    {"team_a": "Динамо",     "team_b": "Шахтар",     "score_a": 1, "score_b": 2},
    {"team_a": "Зоря",       "team_b": "Динамо",     "score_a": 0, "score_b": 1},
    {"team_a": "Шахтар",     "team_b": "Дніпро",     "score_a": 3, "score_b": 1},
    {"team_a": "Динамо",     "team_b": "Ворскла",    "score_a": 2, "score_b": 2},
    {"team_a": "Шахтар",     "team_b": "Зоря",       "score_a": 4, "score_b": 0},
    {"team_a": "Дніпро",     "team_b": "Ворскла",    "score_a": 1, "score_b": 1},
    {"team_a": "Зоря",       "team_b": "Дніпро",     "score_a": 2, "score_b": 1},
    {"team_a": "Ворскла",    "team_b": "Шахтар",     "score_a": 0, "score_b": 3}
]
# Ваш код:

points = {}

for match in matches:
    team_a = match['team_a']
    team_b = match['team_b']
    score_a = match['score_a']
    score_b = match['score_b']

    points[team_a] = points.get(team_a, 0)
    points[team_b] = points.get(team_b, 0)

    if score_a > score_b:
        points[team_a] += 3
    elif score_b > score_a:
        points[team_b] += 3
    else:
        points[team_a] += 1
        points[team_b] += 1

sorted_teams = sorted(points.items(), key=lambda item: item[1], reverse=True)

print("Рейтинг команд:")
for team, score in sorted_teams:
    print(f"{team}: {score} очок")
````

````python
import random

user_score = 0
computer_score = 0

choices = ["камінь", "ножиці", "папір"]

print("Гра 'Камінь-Ножиці-Папір'. Перший до 3 перемог!")

while user_score < 3 and computer_score < 3:
    user_choice = input("\nВиберіть (камінь, ножиці, папір) або 'вихід': ").lower().strip()

    if user_choice == 'вихід':
        print("Дякуємо за гру!")
        break

    if user_choice not in choices:
        print("Некоректний вибір. Будь ласка, введіть 'камінь', 'ножиці' або 'папір'.")
        continue

    computer_choice = random.choice(choices)

    print(f"Ви обрали: {user_choice.capitalize()}")
    print(f"Комп'ютер обрав: {computer_choice.capitalize()}")

    if user_choice == computer_choice:
        print("Нічия!")
    elif (
        (user_choice == "камінь" and computer_choice == "ножиці") or
        (user_choice == "ножиці" and computer_choice == "папір") or
        (user_choice == "папір" and computer_choice == "камінь")
    ):
        print("Ви виграли раунд!")
        user_score += 1
    else:
        print("Комп'ютер виграв раунд!")
        computer_score += 1

    print(f"Рахунок: Ви {user_score} - Комп'ютер {computer_score}")

if user_score == 3:
    print("\nВітаємо! Ви перемогли у грі!")
elif computer_score == 3:
    print("\nКомп'ютер переміг у грі. Спробуйте ще раз!")
````

- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:

balance = 5000
history = []

print("Вітаємо в симуляторі банкомату!")

while True:
    action = input("\nВиберіть дію (баланс, зняти, поповнити, історія, вихід): ").lower().strip()

    if action == "вихід":
        print("Дякуємо за використання банкомату!")
        break

    elif action == "баланс":
        print(f"Ваш поточний баланс: {balance:.2f} грн")

    elif action == "зняти":
        try:
            amount = float(input("Введіть суму для зняття: "))
            if amount <= 0:
                print("Сума для зняття повинна бути додатною.")
            elif amount > balance:
                print("Недостатньо коштів на балансі.")
            else:
                balance -= amount
                history.append({"type": "зняття", "amount": amount})
                print(f"Знято {amount:.2f} грн. Ваш новий баланс: {balance:.2f} грн")
        except ValueError:
            print("Некоректний ввід. Будь ласка, введіть число.")

    elif action == "поповнити":
        try:
            amount = float(input("Введіть суму для поповнення: "))
            if amount <= 0:
                print("Сума для поповнення повинна бути додатною.")
            else:
                balance += amount
                history.append({"type": "поповнення", "amount": amount})
                print(f"Поповнено {amount:.2f} грн. Ваш новий баланс: {balance:.2f} грн")
        except ValueError:
            print("Некоректний ввід. Будь ласка, введіть число.")

    elif action == "історія":
        if history:
            print("\nІсторія операцій:")
            for entry in history:
                print(f"  Тип: {entry['type'].capitalize()}, Сума: {entry['amount']:.2f} грн")
        else:
            print("Історія операцій порожня.")

    else:
        print("Невірна дія. Спробуйте ще раз.")
````

````python
# Ваш код для завдання 46:

tasks = []

print("Вітаємо в менеджері задач!")

while True:
    action = input("\nВиберіть дію (додати, показати, виконати, видалити, вихід): ").lower().strip()

    if action == "вихід":
        print("Дякуємо за використання менеджера задач!")
        break

    elif action == "додати":
        task_description = input("Введіть опис нової задачі: ").strip()
        if task_description:
            tasks.append({"task": task_description, "done": False})
            print(f"Задача '{task_description}' додана.")
        else:
            print("Опис задачі не може бути порожнім.")

    elif action == "показати":
        if not tasks:
            print("Список задач порожній.")
        else:
            print("\nСписок задач:")
            for i, task_item in enumerate(tasks):
                status = "[✓]" if task_item["done"] else "[ ]"
                print(f"{i+1}. {status} {task_item['task']}")

    elif action == "виконати":
        if not tasks:
            print("Список задач порожній. Немає що виконувати.")
            continue
        try:
            task_num = int(input("Введіть номер задачі, яку потрібно відмітити як виконану: "))
            if 1 <= task_num <= len(tasks):
                tasks[task_num - 1]["done"] = True
                print(f"Задача '{tasks[task_num - 1]['task']}' відмічена як виконана.")
            else:
                print("Некоректний номер задачі.")
        except ValueError:
            print("Будь ласка, введіть дійсний номер.")

    elif action == "видалити":
        if not tasks:
            print("Список задач порожній. Немає що видаляти.")
            continue
        try:
            task_num = int(input("Введіть номер задачі, яку потрібно видалити: "))
            if 1 <= task_num <= len(tasks):
                removed_task = tasks.pop(task_num - 1)
                print(f"Задача '{removed_task['task']}' видалена.")
            else:
                print("Некоректний номер задачі.")
        except ValueError:
            print("Будь ласка, введіть дійсний номер.")

    else:
        print("Невірна дія. Будь ласка, виберіть з 'додати', 'показати', 'виконати', 'видалити', 'вихід'.")
````

### Завдання 52. Гра в слова ⭐⭐⭐

Реалізуйте гру: дано стартове слово. Гравець має ввести наступне слово, яке починається на ОСТАННЮ літеру попереднього. Якщо слово вже використовувалось — програш. Якщо слово починається не на ту літеру — програш. Цикл триває, поки гравець не програє або не введе `здаюсь`.

> 💡 **Підказка:** Зберігайте всі використані слова у списку. Першу і останню літеру отримуйте через індекси: `word[0]` і `word[-1]`.

````python
# Ваш код:

print("Вітаємо у грі в слова!\n")

start_word = "програма"
used_words = [start_word]
current_word = start_word

print(f"Перше слово: {start_word.capitalize()}")
print("Правила: кожне наступне слово має починатися на останню літеру попереднього. Слова не можуть повторюватися.\n")

while True:
    print(f"Попереднє слово: {current_word.capitalize()}")
    last_letter = current_word[-1].lower()

    user_input = input(f"Введіть слово, що починається на '{last_letter}' (або 'здаюсь'): ").lower().strip()

    if user_input == 'здаюсь':
        print("Ви здалися. Дякуємо за гру!")
        break

    if not user_input.isalpha():
        print("Некоректний ввід. Будь ласка, вводьте тільки літери.")
        continue

    if user_input in used_words:
        print(f"Програш! Слово '{user_input}' вже було використано.")
        break

    if user_input[0].lower() != last_letter:
        print(f"Програш! Слово '{user_input}' не починається на літеру '{last_letter}'.")
        break

    print(f"Слово '{user_input}' прийнято!")
    used_words.append(user_input)
    current_word = user_input
    print(f"Використані слова: {', '.join(used_words)}\n")

print("Кінець гри.")
````
