## 📖 Розгляд теоретичної частини

Для того, щоб повторити якийсь блок коду **`кілька разів`** або **`повторювати, доки виконується деяка умова`**, у Python реалізовані цикли. Вони є фундаментальною конструкцією, яка дозволяє повторювати виконання певного блоку коду кілька разів.



Існують два основних типи циклів: `for` та `while`


### Цикл `while`

Цикл while часто використовується, коли кількість ітерацій заздалегідь невідома або коли ітерації залежать від змінних, які можуть бути модифіковані в процесі виконання циклу.

Цикл while виконує блок коду, поки задана умова є істинною (True). Як тільки умова стає неправдивою (False), цикл закінчується.




```
while condition:
    # виконувати дії, поки condition є True
```



````python
while True :
  print('Hi')
````

````python
number = 5
while number == 5:
  print(number)
````

````python
number = 5
while number  > 0 :
  number = number - 1
  print(number)
````

````python
number = 5
while number  > 0 :
  number = number - 1
  if number % 2 == 0 :
    continue
  print(number)
````

````python
number = 5
while True :
  number = number - 1
  if number <= 0:
    break
  print(number)
````

### Цикл `for`

Цикл  `for`  перебирає елементи.

````python
for el in 1, 2, 3, 'hi':
  print(el)
````

Функція `range` у поєднанні з циклом `for` у Python є потужним інструментом для контролю повторення дій певну кількість разів. Цей механізм часто використовується для ітерації через послідовність чисел, що робить його особливо корисним для різних задач, від базових до складних алгоритмів.



Функція range створює послідовність чисел. Вона може бути використана різними способами:



`range(stop)`: Створює послідовність чисел від 0 до stop - 1.

`range(start, stop)`: Генерує числа від start до stop - 1.

`range(start, stop, step)`: Створює числа від start до stop - 1, з кроком step.

цикл `for` разом з `range` для ітерації через послідовність чисел виглядає так:

````python
range(5)
````

````python
for el in  range(5):
  print(el)
````

````python
for el in  range(2, 5):
  print(el)
````

````python
for el in  range(2, 15, 3):
  print(el)
````

**Задача 6.1**  
Зчитайте `n` пар «ключ-значення» і збережіть їх у словник. Виведіть значення за заданим ключем.

```
Вхід:   3 / apple 1 / banana 2 / cherry 3 / banana
Вихід:  2
```

````python
n = int(input("n: "))
my_dict = {}
for i in range(n):
   # key, value = input("key, value: ").split() # введення у форматі "key value"
   key = input(f"key{i+1}: ")
   value = input(f"value{i+1}: ")

   my_dict[key] = value

my_dict.get(input("search_key: "), "nan")
````
