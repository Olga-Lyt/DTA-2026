```python
import numpy as np
import pandas as pd

data = {
    "Days": [1, 2, 3, 4, 5, 6, 7],
    "Coins": [2, 4, 6, 8, 10, 12, 14]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Days'], df['Coins'], color='blue')
plt.title("Графік залежності")
plt.xlabel("Days")
plt.ylabel("Coins")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Days"]]
Y = df["Coins"]

model = LinearRegression()
model.fit(X, Y)

# Вивід коефіцієнтів
print("Вага (W):", model.coef_)
print("Зміщення (b):", model.intercept_)
```

```python
# Прогноз
Y_pred = model.predict(X)

plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red')
plt.title("Лінійна регресія")
plt.xlabel("Days")
plt.ylabel("Coins")
plt.show()
```

```python
# Прогноз на Х день
new_x = [[10]]
predict_coins = model.predict(new_x)
print(f"Прогноз на {new_x[0][0]} день: {predict_coins[0]} монеток")
```
