Типи результатів функцій

```python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
```

```python
# Повертає NonType без зміни даних і виводить результат
def info(d):
    print(d)

type(info(my_dict))
```

```python

# Повертає NonType змінює початкові дані і виводить результат
def info(d):
    d[6] = "six"
    print(d)
print(f"my_dict: {my_dict}")
type(info(my_dict))
```

```python
# Повертає NonType змінює початкові дані і НЕ виводить результат
def add_el(d):
    d[6] = "six"

print(type(add_el(my_dict)))
print(f"my_dict: {my_dict}")
```
