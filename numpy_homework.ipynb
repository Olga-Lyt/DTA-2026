## Розділ 3 — Математичні операції та векторизація

> **Що повторюємо:** арифметика з масивами, `np.sqrt()`, `np.round()`,
> агрегатні функції `.sum()`, `.mean()`, `.std()`, `.min()`, `.max()`, `np.median()`

### ✏️ Завдання 3.1

Є масив цін товарів у доларах. Виконай **без циклів**:

```python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])
```

1. Переведи ціни в гривні (курс: **1 USD = 41.5 UAH**), результат округли до 2 знаків
2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
5. Виведи результати з поясненнями

````python
import numpy as np

prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])

# Твій код:
exchange_rate = 41.5
prices_uah = np.round(prices_usd * exchange_rate, 2)
print(f"Ціни в гривнях: {prices_uah}")

vat_rate = 1.2
prices_with_vat = np.round(prices_uah * vat_rate, 2)
print(f"Ціни з ПДВ: {prices_with_vat}")

min_price_vat = np.min(prices_with_vat)
max_price_vat = np.max(prices_with_vat)
mean_price_vat = np.round(np.mean(prices_with_vat), 2)
print(f"Мінімальна ціна з ПДВ: {min_price_vat} UAH")
print(f"Максимальна ціна з ПДВ: {max_price_vat} UAH")
print(f"Середня ціна з ПДВ: {mean_price_vat} UAH")

total_cart_value = np.round(np.sum(prices_with_vat), 2)
print(f"Загальна вартість кошика з ПДВ: {total_cart_value} UAH")
````

### ✏️ Завдання 3.2

Є дані про кількість товарів на складі та їх собівартість.
Виконай **без циклів**:

```python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']
```

1. Порахуй загальну вартість кожного товару на складі (`quantity * cost_each`)
2. Знайди загальну вартість всього складу
3. Знайди товар з найбільшою сумарною вартістю — використай `np.argmax()`
4. Знайди товар з найменшою сумарною вартістю — `np.argmin()`

````python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']

# Твій код:

total_cost_per_item = quantity * cost_each
print(f"Загальна вартість кожного товару: {total_cost_per_item}")

total_stock_value = np.sum(total_cost_per_item)
print(f"Загальна вартість всього складу: {total_stock_value} UAH")

max_value_index = np.argmax(total_cost_per_item)
max_value_item_name = names[max_value_index]
print(f"Товар з найбільшою сумарною вартістю: {max_value_item_name} ({total_cost_per_item[max_value_index]} UAH)")

min_value_index = np.argmin(total_cost_per_item)
min_value_item_name = names[min_value_index]
print(f"Товар з найменшою сумарною вартістю: {min_value_item_name} ({total_cost_per_item[min_value_index]} UAH)")

mean_item_value = np.round(np.mean(total_cost_per_item), 2)
print(f"Середня вартість одного найменування на складі: {mean_item_value} UAH")
````

### ✏️ Завдання 4.1

Є масив часу виконання замовлень у хвилинах (норма — до 30 хв):

```python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])
```

1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)
2. Знайди всі **запізнілі** замовлення (> 30 хв)
3. Порахуй кількість вчасних і запізнілих
4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
5. Яка середня затримка серед запізнілих замовлень?

````python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])

# Твій код:

on_time_orders = delivery_time[delivery_time <= 30]
print(f"Замовлення, виконані вчасно: {on_time_orders}")

late_orders = delivery_time[delivery_time > 30]
print(f"Запізнілі замовлення: {late_orders}")

num_on_time = on_time_orders.size
num_late = late_orders.size
print(f"Кількість вчасних замовлень: {num_on_time}")
print(f"Кількість запізнілих замовлень: {num_late}")

total_orders = delivery_time.size
percentage_late = np.round((num_late / total_orders) * 100, 1)
print(f"Відсоток запізнілих замовлень: {percentage_late}%")

if num_late > 0:
    average_delay = np.round(np.mean(late_orders - 30), 1)
    print(f"Середня затримка серед запізнілих замовлень: {average_delay} хв")
else:
    print("Запізнілих замовлень немає.")
````

### ✏️ Завдання 5.1

Є 1D масив з 24 показників продажів (по годинах за добу):

```python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])
```

1. Перетвори масив у матрицю **6×4** (6 блоків по 4 години) і виведи
2. Знайди суму продажів у кожному 4-годинному блоці (по рядках)
3. Відсортуй вихідний 1D масив за **зростанням** і виведи
4. Знайди **індекси** годин відсортованих за зростанням продажів — `np.argsort()`
5. Виведи топ-3 найпродуктивніших години (їхні номери, рахуючи з 0)

````python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])

# Твій код:

hourly_matrix = hourly.reshape(6, 4)
print(f"Матриця 6x4 (продажі по годинах):\n{hourly_matrix}")

block_sums = np.sum(hourly_matrix, axis=1)
print(f"\nСума продажів у кожному 4-годинному блоці: {block_sums}")

sorted_hourly = np.sort(hourly)
print(f"\nВідсортований 1D масив (зростання): {sorted_hourly}")

sorted_indices = np.argsort(hourly)
print(f"\nІндекси годин відсортованих за зростанням продажів: {sorted_indices}")


top_3_indices = np.argsort(hourly)[-3:]

top_3_hours = top_3_indices + 1
print(f"\nТоп-3 найпродуктивніших години (номер години, рахуючи з 0): {top_3_indices}")
print(f"\nТоп-3 найпродуктивніших години (номер години, рахуючи з 1): {top_3_hours}")
````

### ✏️ Завдання 6.1

Змоделюй результати маркетингового опитування 50 клієнтів.

1. Встанови `np.random.seed(99)`
2. Згенеруй масив `ages` — вік 50 клієнтів: цілі числа від 18 до 65
3. Згенеруй масив `scores` — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
4. Обріж `scores` так щоб значення були від 1 до 10 — `np.clip(arr, 1, 10)`
5. Виведи:
   - середній вік клієнтів
   - середню оцінку задоволеності
   - кількість клієнтів з оцінкою >= 8 («задоволені»)
   - відсоток задоволених

````python
np.random.seed(99)

ages = np.random.randint(18, 66, size=50)
print(f"Масив віку клієнтів (перші 10): {ages[:10]}...")

scores = np.round(np.random.normal(loc=7, scale=1.5, size=50), 1)
print(f"\nМасив оцінок задоволеності (без обрізки, перші 10): {scores[:10]}...")

scores_clipped = np.clip(scores, 1, 10)
print(f"Масив оцінок задоволеності (з обрізкою, перші 10): {scores_clipped[:10]}...")

mean_age = np.mean(ages)
print(f"\nСередній вік клієнтів: {mean_age:.2f}")

mean_score = np.mean(scores_clipped)
print(f"Середня оцінка задоволеності: {mean_score:.2f}")

satisfied_clients = scores_clipped[scores_clipped >= 8]
num_satisfied = satisfied_clients.size
print(f"Кількість клієнтів з оцінкою >= 8: {num_satisfied}")

percentage_satisfied = np.round((num_satisfied / scores_clipped.size) * 100, 1)
print(f"Відсоток задоволених клієнтів: {percentage_satisfied}%")
````

## 🏆 Фінальний проєкт — Аналіз даних магазину

> **Використовуємо все з тьюторіалу разом**

Ти аналітик даних невеликого магазину електроніки.
Тобі передали масив продажів за 8 тижнів (56 днів).

Виконай повний аналіз та дай відповіді на запитання керівника.

````python
# Дані — не змінюй цю клітинку
np.random.seed(2024)
base = np.random.randint(200, 800, 56)

# Вихідні продаються краще
weekend_boost = np.array([1 if i % 7 in [5, 6] else 0 for i in range(56)])
sales_56 = (base + weekend_boost * np.random.randint(100, 300, 56)).astype(int)

print('Дані отримано. Масив sales_56 готовий до аналізу.')
print(f'Елементів: {sales_56.size}, форма: {sales_56.shape}')
````

### Запитання 1 — Загальна статистика (Розділ 1, 3)
Виведи:
- Загальна виручка за 56 днів
- Середньоденна виручка (округли до 2 знаків)
- Мінімальний та максимальний денний продаж
- Стандартне відхилення (наскільки продажі «скачуть»)

````python
# Твій код:

total_revenue = np.sum(sales_56)
print(f"Загальна виручка за 56 днів: {total_revenue} грн")

daily_average_revenue = np.round(np.mean(sales_56), 2)
print(f"Середньоденна виручка: {daily_average_revenue} грн")

min_daily_sales = np.min(sales_56)
max_daily_sales = np.max(sales_56)
print(f"Мінімальний денний продаж: {min_daily_sales} грн")
print(f"Максимальний денний продаж: {max_daily_sales} грн")

sales_std_dev = np.round(np.std(sales_56), 2)
print(f"Стандартне відхилення продажів: {sales_std_dev} грн")
````

````python
# Твій код:

sales_matrix = sales_56.reshape(8, 7)
print(f"Матриця продажів 8x7:\n{sales_matrix}")

weekly_revenue = np.sum(sales_matrix, axis=1)
print(f"\nВиручка по тижнях: {weekly_revenue} грн")

best_week_index = np.argmax(weekly_revenue)
best_week_number = best_week_index + 1
best_week_revenue = weekly_revenue[best_week_index]
print(f"\nНайкращий тиждень: Тиждень {best_week_number} з виручкою {best_week_revenue} грн")

daily_average_per_weekday = np.round(np.mean(sales_matrix, axis=0), 2)
print(f"\nСередньоденна виручка по днях тижня (Пн-Нд): {daily_average_per_weekday} грн")

most_profitable_day_index = np.argmax(daily_average_per_weekday)
print(f"\nНайприбутковіший день тижня (індекс, 0=Пн): {most_profitable_day_index}")
````

### Запитання 3 — Фільтрація та маскування (Розділ 4)
1. Знайди кількість **«слабких» днів** — продажі нижче 400 грн
2. Знайди кількість **«сильних» днів** — продажі вище 700 грн
3. Яка середня виручка у слабкі дні і яка у сильні?
4. Створи масив `labels` за допомогою `np.where()`:
   - `'Слабкий'` якщо < 400
   - `'Сильний'` якщо > 700
   - `'Нормальний'` в решті випадків (підказка: вкладений `np.where`)
5. Виведи перші 14 днів: продажі та їх мітки

````python
# Твій код:

weak_days_mask = sales_56 < 400
num_weak_days = np.sum(weak_days_mask)
print(f"Кількість «слабких» днів (продажі < 400 грн): {num_weak_days}")

strong_days_mask = sales_56 > 700
num_strong_days = np.sum(strong_days_mask)
print(f"Кількість «сильних» днів (продажі > 700 грн): {num_strong_days}")

if num_weak_days > 0:
    average_weak_sales = np.round(np.mean(sales_56[weak_days_mask]), 2)
    print(f"Середня виручка у слабкі дні: {average_weak_sales} грн")
else:
    print("Немає слабких днів для розрахунку середньої виручки.")

if num_strong_days > 0:
    average_strong_sales = np.round(np.mean(sales_56[strong_days_mask]), 2)
    print(f"Середня виручка у сильні дні: {average_strong_sales} грн")
else:
    print("Немає сильних днів для розрахунку середньої виручки.")

labels = np.where(sales_56 < 400, 'Слабкий',
                  np.where(sales_56 > 700, 'Сильний', 'Нормальний'))


print("\nПерші 14 днів: продажі та їх мітки:")
for i in range(14):
    print(f"День {i+1}: Продажі = {sales_56[i]}, Мітка = {labels[i]}")
````
